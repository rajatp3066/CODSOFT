```python
#import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset
credit_card= pd.read_csv('creditcard.csv')
```

```python
#check distribution of legit transcation & fraud transactions
credit_card['Class'].value_counts()
```

o represent legit transaction and 1 represent fraud transcations

```python
#seperating data for anaysis
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
#Statistical measurement 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values of both transaction by groping 
credit_card.groupby('Class').mean()
```

```python
#Dealing with unbalanced data- undersampling
legit_sample = legit.sample(n = 492)
```

```python
#concantenanting data frame
new_data = pd.concat([legit_sample, fraud], axis =0)
```

axis = 0 means that 492 sample present row wise

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

Data Split into feature and tagrgets

```python
X= new_data.drop(columns = 'Class', axis = 1)
Y= new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split into train & test

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##### Model Trainig (LogisticRegression)

```python
model= LogisticRegression()
```

```python
#for train data

model.fit(X_train, Y_train)
```

##### Model Evaluation

```python
#Accuracy Score
 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```

In conclusion, our credit card fraud detection project has yielded impressive results, with a training data accuracy of 92% and a test data accuracy of 90%.
